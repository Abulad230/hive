```python
import pandas as pd

# Load existing metrics
df = pd.read_csv('C:/Users/lastw/Downloads/UND_hive-2.0.0_processed.csv')
df.head()

```

```python
import os
import subprocess

# Function to run shell commands
def run_command(command):
    result = subprocess.run(command, stdout=subprocess.PIPE, shell=True, text=True)
    return result.stdout.strip()

# Function to extract metrics using Python instead of awk
def extract_git_metrics(file_path):
    log_output = run_command(f"git log --follow --pretty=tformat: --numstat {file_path}")
    added_lines = 0
    deleted_lines = 0
    for line in log_output.split('\n'):
        if line:
            added, deleted, _ = line.split('\t')
            added_lines += int(added)
            deleted_lines += int(deleted)
    return added_lines, deleted_lines

# Initialize new columns
df['lines_added'] = 0
df['lines_deleted'] = 0

# Iterate over DataFrame and extract metrics
for index, row in df.iterrows():
    file_path = row['FileName']
    added, deleted = extract_git_metrics(file_path)
    df.at[index, 'lines_added'] = added
    df.at[index, 'lines_deleted'] = deleted

# Display the updated DataFrame
df.head()
```
